```python
from numpy import *
import control
from control.matlab import *    
import matplotlib.pyplot as plt
```

```python
m = 1
c = 0.2
A = matrix([[0, 1],[0, -c/m]])
B = matrix([[0], [1/m]])
C = matrix([[1,0], [0,1]])
D = matrix([[0], [0]])
Q = matrix([[1, 0],[0, 1]])
R = matrix([[0.01]])
```

```python
K, S, E = lqr(A, B, Q, R)
```

```python
sys = control.StateSpace(A - B*K,B*K*matrix([[1],[0]]),C,D)
```

```python
Y, X = step(sys)
```

```python
plt.plot(X,Y)
```
